```python
import pandas as pd
import matplotlib.pyplot as plt


# reading the database
data = pd.read_csv("Heart Attack Data Set.csv")

# Scatter plot with day against tip
plt.scatter(data['age'], data['trestbps'])

# Adding Title to the Plot
plt.title("Scatter Plot")

# Setting the X and Y labels
plt.xlabel('age')
plt.ylabel('trestbps')

plt.colorbar()

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt


# reading the database
data = pd.read_csv("Heart Attack Data Set.csv")

# Scatter plot with day against tip
plt.plot(data['fbs'])
plt.plot(data['chol'])

# Adding Title to the Plot
plt.title("Scatter Plot")

# Setting the X and Y labels
plt.xlabel('fbs')
plt.ylabel('chol')

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt


# reading the database
data = pd.read_csv("Heart Attack Data Set.csv")

# Bar chart with day against tip
plt.bar(data['sex'], data['cp'], color= ['red','orange'])
# c= ['red','orange']

plt.title("Bar Chart")

# Setting the X and Y labels
plt.xlabel('sex')
plt.ylabel('cp')

# Adding the legends
plt.show()
```
